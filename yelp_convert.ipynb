```python
import os
import sys
import csv
import json
```

```python
def convert_to (filename) :
    
    if "review" in filename :
        
        output_file = open("yelp_review.csv",'w')
        fields = [ "user_id","business_id", "stars"]
        
        with_fields = csv.DictWriter(output_file, fieldnames = fields, lineterminator='\n')
        print ("Converting " + filename + " to " + "yelp_review.csv ...")
        
        for line in  open(filename,'r') :
            r = json.loads(line)
            with_fields.writerow({'user_id': r['user_id'],'business_id': r['business_id'],'stars': r['stars'] })
            
    if "user" in filename:
        
        output_file = open("yelp_user.csv",'w')
        fields = ["user_id", "name"]
        
        with_feilds = csv.DictWriter(output_file,fields,lineterminator = '\n')
        print ("Converting " + filename + " to " + "yelp_review.csv ...")
        
        
        for line in open(filename,'r'):
            r = json.loads(line)
            
            n = r["name"]
            n1 = n.encode("ascii","ignore")
            with_feilds.writerow({"user_id":r["user_id"],"name":n1})
            
    if "business" in filename:
        
        output_file = open("yelp_business.csv","w")
        feilds = ["business_id","city", "name" ,"categories", "review_count", "stars"]
        
        with_feilds = csv.DictWriter(output_file,feilds,lineterminator="\n")
        print ("Converting " + filename + " to " + "yelp_review.csv ...")
        
        
        for line in open(filename,"r"):
            r = json.loads(line)
            
            names = str(r['categories'])
            if "Restaurant" in names:
                n = r["name"]
                n1 = n.encode("ascii","ignore")
                c = r["city"]
                c1 = c.encode("ascii","ignore")
                with_feilds.writerow({"business_id":r["business_id"],"city": c1,"name":n1 ,"categories":r["categories"] ,"review_count":r["review_count"], "stars":r["stars"]})
                
    
    else:
        print("no input file")
        exit()
                                       
    
def maine():
    args = sys.argv[1:]
    if not args:
        print ('usage: file')
        sys.exit(1)        
    convert_to(sys.argv[1])
                
        
        
if __name__ == '__main__':
    maine()                                              
        
```
